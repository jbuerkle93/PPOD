```python
### 3.3 testing around with the clustering algorithm "Anatomize" by Xiao and Tao 2006
#-> i found a python 2.7 version, which unfortunately does not work with my python version 3
import pandas as pd
import random

dataset = pd.read_csv('Student Mental Health.csv')

def anatomize_algorithm(dataset, l):
    QIT = pd.DataFrame(columns=['ID', 'Choose your gender', 'Your current year of Study', 'Marital status', 'What is your course?','What is your CPGA', 'GroupID'])
    ST = pd.DataFrame(columns=['GroupID', 'As value', 'Count'])
    gcnt = 0

    while True:
        # Hash the tuples by 'Choose your gender' (assuming it's the quasi-identifier As)
        hash_buckets = dataset.groupby('Choose your gender', group_keys=False).apply(lambda x: x.sample(frac=1)).reset_index(drop=True)

        # Check if there are at least l non-empty hash buckets
        non_empty_buckets = [bucket for _, bucket in hash_buckets.groupby('Choose your gender') if not bucket.empty]
        if len(non_empty_buckets) < l:
            break

        # Group-creation step
        gcnt += 1
        QIgcnt = pd.DataFrame(columns=dataset.columns)

        # Select the set of l largest buckets
        S = sorted(non_empty_buckets, key=lambda x: len(x), reverse=True)[:l]

        for bucket in S:
            # Remove an arbitrary tuple from the bucket
            t = bucket.sample(1)
            bucket.drop(t.index, inplace=True)

            # Add tuple to the current QI-group
            QIgcnt = pd.concat([QIgcnt, t])

        # Residue-assignment step
        for _, bucket in hash_buckets.groupby('Choose your gender'):
            if len(bucket) == 1:
                t = bucket.iloc[0]
                S_prime = [j for j in range(1, gcnt + 1) if t['Choose your gender'] not in QIgcnt['Choose your gender'].values]

                if S_prime:
                    # Assign t to a random QI-group in S'
                    random_group = random.choice(S_prime)
                    QIgcnt = pd.concat([QIgcnt, t])

        # Populate QIT and ST
        for j in range(1, gcnt + 1):
            QIT = pd.concat([QIT, QIgcnt.assign(GroupID=j)])

            for v in QIgcnt['Choose your gender'].unique():
                cj_v = QIgcnt['Choose your gender'].value_counts().get(v, 0)
                ST = pd.concat([ST, pd.DataFrame({'GroupID': [j], 'As value': [v], 'Count': [cj_v]})], ignore_index=True)

    return QIT, ST

# Example usage
l_value = 2  # You can adjust the value of l
QIT_result, ST_result = anatomize_algorithm(dataset, l_value)
print("Quasi-Identifiers Table:")
print(QIT_result)
print("\nSensitive Attribute Table:")
print(ST_result)
```

```python
# copying the python 2.7 package and trying to run it in python 3

import pandas as pd
import random
import heapq

_DEBUG = True

class SABucket:
    def __init__(self, value, index):
        self.value = value
        self.member = []
        self.index = index

    def __lt__(self, other):
        return self.value < other.value

    def pop_element(self):
        # Check if the bucket is not empty before popping
        if self.member:
            return self.member.pop()
        else:
            # If the bucket is empty, return None or handle it as needed
            return None

    def __len__(self):
        return len(self.member)

    def add_element(self, record, index):
        # Check if the record is not None before appending
        if record is not None:
            self.member.append(record[:])
            self.checklist.add(index)


class Group:
    def __init__(self):
        self.index = 0
        self.member = []
        self.checklist = set()

    def add_element(self, record, index):
        self.member.append(record[:])
        self.checklist.add(index)

    def check_index(self, index):
        return index in self.checklist

    def __len__(self):
        return len(self.member)


def build_SA_bucket(data):
    buckets = {}
    bucket_heap = []

    for i, record in enumerate(data):
        sa_value = record[-1]
        try:
            buckets[sa_value].append(record)
        except KeyError:
            buckets[sa_value] = [record]

    for key in buckets.keys():
        random.shuffle(buckets[key])

    for i, bucketed_record in enumerate(buckets.values()):
        length = len(bucketed_record) * -1
        if length == 0:
            continue
        heapq.heappush(bucket_heap, (length, SABucket(bucketed_record, i)))

    return buckets, bucket_heap


def assign_to_groups(buckets, bucket_heap, L):
    groups = []
    print("Assigning records to groups...")
    while len(bucket_heap) >= L:
        newgroup = Group()
        length_list = []
        SAB_list = []

        for i in range(L):
            length, bucket = heapq.heappop(bucket_heap)
            length_list.append(length)
            SAB_list.append(bucket)

        for i in range(L):
            bucket = SAB_list[i]
            length = length_list[i]
            newgroup.add_element(bucket.pop_element(), bucket.index)
            length += 1
            if length == 0:
                continue
            heapq.heappush(bucket_heap, (length, bucket))

        groups.append(newgroup)
        print(f"Group {len(groups)} has {len(newgroup)} records.")

    return groups


def residue_assign(groups, bucket_heap):
    suppress = []
    print("Handling residue records...")
    while len(bucket_heap):
        _, bucket = heapq.heappop(bucket_heap)
        index = bucket.index
        candidate_set = []

        for group in groups:
            if not group.check_index(index):
                candidate_set.append(group)

        if len(candidate_set) == 0:
            suppress.extend(bucket.member[:])

        while bucket.member:
            candidate_len = len(candidate_set)
            if candidate_len == 0:
                break

            current_record = bucket.pop_element()
            group_index = random.randrange(candidate_len)
            group = candidate_set.pop(group_index)
            group.add_element(current_record, index)

        if len(bucket) >= 0:
            suppress.extend(bucket.member[:])

    print(f"Number of suppressed records: {len(suppress)}")

    return groups, suppress


def split_table(groups):
    qi_table = []
    sa_table = []
    result = []

    for i, group in enumerate(groups):
        group.index = i
        result.append(group.member[:])

        for record in group.member:
            qi_part = record[:-1]
            qi_part.append(i)
            sa_part = [record[-1]]
            sa_part.insert(0, i)
            qi_table.append(qi_part)
            sa_table.append(sa_part)

    return qi_table, sa_table, result


def anatomize(data, L):
    if _DEBUG:
        print('*' * 10)
        print("Begin Anatomizer!")
    print("L=%d" % L)

    buckets, bucket_heap = build_SA_bucket(data)
    print("Number of SA Buckets:", len(buckets))

    groups = assign_to_groups(buckets, bucket_heap, L)
    print("Number of Groups:", len(groups))

    groups, suppress = residue_assign(groups, bucket_heap)

    qi_table, sa_table, result = split_table(groups)

    if _DEBUG:
        for i in range(len(qi_table)):
            print(qi_table[i] + sa_table[i])

    return result


# Example usage
df3 = pd.read_csv('Student Mental Health.csv')
df3.loc[:, 'Age'] = df3['Age'].astype(str)

RESULT = anatomize(df3.values.tolist(), 2)
```
