```python
### 3.1.2 ###

import pandas as pd
import spacy

# Sample DataFrame
df312 = pd.read_csv('tweet_emotions.csv')

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Function to identify PII entities
def identify_pii(text):
    doc = nlp(text)
    pii_entities = []
    for ent in doc.ents:
        if ent.label_ in ['PERSON', 'DATE', 'EVENT', 'GPE', 'ORG']:
            pii_entities.append((ent.text, ent.label_))
    return pii_entities

# Apply the function to the DataFrame
df312['PII_Entities'] = df312['content'].apply(identify_pii)


# Display the DataFrame with identified PII entities
print(df312)



```

```python
# Initialize counts for each entity type
pii_counts = {'PERSON': 0, 'DATE': 0, 'EVENT': 0, 'GPE': 0, 'ORG': 0}

# Apply the function to the DataFrame and aggregate counts
for text in df312['content']:
    entities = identify_pii(text)  # Assuming identify_pii is already defined and returns a list of tuples (entity, label)
    for entity, label in entities:
        if label in pii_counts:
            pii_counts[label] += 1

# Display counts of PII entities by entity type
print("Counts of PII Entities by Type:")
for entity_type, count in pii_counts.items():
    print(f"{entity_type}: {count}")
```

```python
# 3.1.3 algorithm 

import pandas as pd
import re
import random
import string

df313 = df312.copy(deep=True)

def replace_twitter_handles(text):
    # Define regular expression pattern for detecting Twitter handles
    twitter_handle_pattern = r'(?<!\w)@(\w{1,15})\b'

    # Replace Twitter handles with the generic handle
    replaced_text = re.sub(twitter_handle_pattern, r'@hidden_user_name', text)

    return replaced_text

# randomize tweet_id
def randomize_id(number):
    digits = [int(d) for d in str(number)]
    random.shuffle(digits)
    return int(''.join(map(str, digits)))

df313['tweet_id'] = df313['tweet_id'].apply(randomize_id)

# Create a copy of the 'content' column
df313['modified_content'] = df313['content']


#Randomize person names
# Function to generate a random string
def random_string(length):
    letters = string.ascii_letters
    return ''.join(random.choice(letters) for _ in range(length))

# Function to replace PERSON, DATE, ORG, and GPE entities with appropriate values
def replace_entities(row):
    content = row['content']
    pii_entities = row['PII_Entities']
    
    # Replace each entity based on its label
    for ent_text, ent_label in pii_entities:
        if ent_label == "PERSON":
            # Check if the identified PERSON is a Twitter handle
            if ent_text.startswith('@') and content.find(f' {ent_text}') != -1:
                continue  # Skip replacing if it's a Twitter handle
            # Generate meaningful names (you can replace this with your own logic)
            first_names = [
                "John", "Emma", "Michael", "Sophia", "James", "Olivia", "William", "Isabella", 
                "Alexander", "Emily", "Jacob", "Charlotte", "Benjamin", "Amelia", "Daniel", "Mia",
                "Ethan", "Harper", "Matthew", "Ella", "Logan", "Abigail", "David", "Ava",
                "Samuel", "Scarlett", "Joseph", "Grace", "Gabriel", "Lily", "Jackson", "Chloe"
            ]
            replacement = random.choice(first_names)
            content = content.replace(ent_text, replacement)
        elif ent_label == "DATE":
            # Replace DATE entity with "hidden_date"
            content = re.sub(r'\b{}\b'.format(re.escape(ent_text)), "hidden_date", content)
        elif ent_label == "ORG":
            # Replace ORG entity with "hidden_org"
            content = re.sub(r'\b{}\b'.format(re.escape(ent_text)), "hidden_org", content)
        elif ent_label == "GPE":
            # Replace GPE entity with "hidden_GPE"
            content = re.sub(r'\b{}\b'.format(re.escape(ent_text)), "hidden_GPE", content)
    
    return content


# Apply the function to replace PERSON entities in the 'content' column
#df313['content'] = df313.apply(replace_person_entities, axis=1)


# Apply the function to replace PERSON entities in the copied 'modified_content' column
df313['modified_content'] = df313.apply(replace_entities, axis=1)


# Replace Twitter handles in the 'content' column
df313['modified_content'] = df313['modified_content'].apply(replace_twitter_handles)

# Detecting links starting with 'http'
link_pattern = r'\bhttps?://\S+\b'

# Replace links with the generic link in the 'content' column
df313['modified_content'] = df313['modified_content'].str.replace(link_pattern, 'http://hidden_link', regex=True)



# Display the modified DataFrame
#print(df313)
filtered_df = df313[df313['PII_Entities'].apply(lambda x: any(label == "PERSON" for _, label in x))]

# Print the filtered DataFrame
print(filtered_df)
```

```python
# Step 2 - Prepare the data for better analysis

from scipy import stats

# transform data for better analysis
# race, gender,

# Define a threshold for z-score
threshold = 3

# Define the mapping from gender to numerical values
gender_mapping = {'M': 1, 'F': 2}

# Transform the gender column using the mapping
QI['gender'] = QI['gender'].map(gender_mapping)

race_mapping = {
    'A': 0,  # Asian
    'B': 1,  # Black
    'H': 2,  # Hispanic
    'N': 3,  # Native
    'O': 4,  # Other
    'W': 5   # White
}

# Define the mapping for state categories to numerical codes
state_mapping = {state: i for i, state in enumerate(QI['state'].unique())}

# Define the mapping for manner_of_death categories to numerical codes
manner_of_death_mapping = {
    'shot': 0,
    'shot and Tasered': 1
}

# Define the mapping for flee categories to numerical codes
flee_mapping = {
    'Not fleeing': 0,
    'Car': 1,
    'Foot': 2,
    'Other': 3
}


# Replace the race column values with numerical codes
QI['race'] = QI['race'].replace(race_mapping)

# Replace the gender column values with boolean values
QI['gender'] = QI['gender'].replace(gender_mapping)

# Replace the state column values with numerical codes
QI['state'] = QI['state'].replace(state_mapping)

# Replace the manner_of_death column values with numerical codes
QI['manner_of_death'] = QI['manner_of_death'].replace(manner_of_death_mapping)

# Replace the flee column values with numerical codes
QI['flee'] = QI['flee'].replace(flee_mapping)





# remove outliers


# Select numeric columns for outlier detection
numeric_columns = QI.select_dtypes(include=['number']).columns

# Calculate z-scores for each numeric column
z_scores = stats.zscore(QI[numeric_columns])

# Find outliers
outliers = (z_scores > threshold).any(axis=1)

# Remove rows containing outliers
QI_no_outliers = QI[~outliers]

# Print the DataFrame without outliers
print(QI_no_outliers)
print(QI)
print(outliers)
```
